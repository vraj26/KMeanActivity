```python
#first import neccessary libraries
import pandas as pd #great for data tables
import numpy as np #great for maniopulating data
import matplotlib.pyplot as plt #great for plotting data and/or visualizing
from sklearn.cluster import KMeans
```

A Standard K-means clustering example

```python
#reading dataset
df = pd.read_csv("faithful.csv")
df
```

```python
#visualizing the data
plt.scatter(df['eruption'], df['waiting'])
plt.ylabel('Time between Eruptions (mins)')
plt.xlabel('Duration of Eruption (mins)')
plt.title('Eruptions from Old Faithful Geyser')
```

```python
#convert dat in features
X = df.values
X
```

```python
#build K-means object
kmeans = KMeans(n_clusters = 2,
                init = 'random',
                )

#fits algo onto our data
kmeans.fit(X)
```

The result of our fit

```python
#category 1 and 0 eruptions
output = kmeans.labels_
output
```

```python
#cateogrizing the outputs
category0 = df[output == 0]
category1 = df[output == 1]


#plotting data with new categories
plt.scatter(category0['eruption'], category0['waiting'], label='category0')
plt.scatter(category1['eruption'], category1['waiting'], label='category1')

plt.legend()
plt.ylabel('Time between Eruptions (mins)')
plt.xlabel('Duration of Eruption (mins)')
plt.title('Eruptions from Old Faithful Geyser')
```

This eruption lasts for 3 minutes and came 75.2 minutes adter previous eruption

```python
new_eruption = 3
new_waiting = 75.2

kmeans.predict([[new_eruption, new_waiting]])
```

A non-spherical slutering example

```python

#Output are location of the darts
df = pd.read_csv('bulls_eye.csv')[['x', 'y']]
df
```

```python
#visualizing data
plt.scatter(df['x'], df['y'])
plt.title("Bullseye hits")
```

```python
kmeans = KMeans(n_clusters=3, # we have three target regions
                init = 'random',
                max_iter = 300,
                )


kmeans.fit(df.values)


flags = kmeans.labels_
category0 = df[flags == 0] # creating a mask that filters for only category 0
category1 = df[flags == 1] # creating a mask that filters for only category 1
category2 = df[flags == 2] # creating a mask that filters for only category 2

# plotting the categories
plt.scatter(category0['x'], category0['y'], label = 'Category 0')
plt.scatter(category1['x'], category1['y'], label = 'Category 1')
plt.scatter(category2['x'], category2['y'], label = 'Category 2')

plt.title('Bulls Eye')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc = 'lower right')
```

```python
#Convert data into rectangular case to take advantage of KMeans clustering algorithms
#Polar Coordinate conversion

#radius feature
df['r'] = (df['x']**2 + df['y']**2)**0.5

#angle feature
df['phi'] = np.arcsin(df['y']/df['r'])
```

```python
#SAME CODE BUT PLOTTING WITH POLAR 
#try again with kmeans algo

kmeans = KMeans(n_clusters=3, # we have three target regions
                init = 'random',
                max_iter = 400,
                )

kmeans.fit(df[['r', 'phi']].values)

flags = kmeans.labels_
category0 = df[flags == 0] # creating a mask that filters for only category 0
category1 = df[flags == 1] # creating a mask that filters for only category 1
category2 = df[flags == 2] # creating a mask that filters for only category 2

# plotting the categories
plt.scatter(category0['r'], category0['phi'], label = 'Category 0')
plt.scatter(category1['r'], category1['phi'], label = 'Category 1')
plt.scatter(category2['r'], category2['phi'], label = 'Category 2')

plt.title('Bulls Eye')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc = 'lower right')
```

```python
#Output does show the "rectangular" shape when printed with polar coordinates
#A lot easier to use KMeans clustering

#try again with kmeans algo

kmeans = KMeans(n_clusters=3, # we have three target regions
                init = 'random',
                max_iter = 400,
                )

kmeans.fit(df[['r', 'phi']].values)

flags = kmeans.labels_
category0 = df[flags == 0] # creating a mask that filters for only category 0
category1 = df[flags == 1] # creating a mask that filters for only category 1
category2 = df[flags == 2] # creating a mask that filters for only category 2

# plotting the categories
plt.scatter(category0['x'], category0['y'], label = 'Category 0')
plt.scatter(category1['x'], category1['y'], label = 'Category 1')
plt.scatter(category2['x'], category2['y'], label = 'Category 2')

plt.title('Bulls Eye')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc = 'lower right')
```

```python
len(df)

#try to see how many points on the board

#category 2 -15 points
#category 1 - 10 points
#category0 - 5 points

score = 0 # our score will start at zero and increment as we accumulate hits

for label in kmeans.labels_:
    if label == 0: # Category 0
        score += 5
    elif label == 1: # Category 1
        score += 10
    elif label == 2: # Category 2
        score += 15

print(score)
```
